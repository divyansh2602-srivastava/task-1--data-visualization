```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"c:\Users\divya\OneDrive\Documents\age_gender.csv.zip")
df
```

```python
plt.hist(df['age'],color='Red',edgecolor='Black',alpha=0.9,bins=10,histtype='bar')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Gender')
plt.show()
```

Creating a bar chart to show the frequency of each gender in the Dataset

```python
gender_counts = df['gender'].value_counts()
gender_counts.plot(kind='bar',color=['blue','pink'])
plt.title('Gender Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
